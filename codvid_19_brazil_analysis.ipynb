```
import pandas as pd
import bar_chart_race as bcr
from IPython.display import HTML
```

```
df = pd.read_csv("/content/arquivo_geral.csv", sep = ";")
df
```

```
clean_df = df[['data', 'estado', 'casosAcumulados']]
clean_df



# df_pivot_new_cases = df[['data', 'regiao', 'casosNovos']].pivot(index='data', columns='regiao', values='casosNovos')
# df_pivot_new_cases
```

```
df_pivot_total_cases = clean_df.pivot(index='data', columns='estado', values='casosAcumulados')
df_pivot_total_cases.tail(5)
```

```
bcr_html = bcr.bar_chart_race(
    df=df_pivot_total_cases,
    filename= None,
    orientation='h',
    sort='desc',
    label_bars=True,
    use_index=True,
    steps_per_period=10,
    period_length=500,
    figsize=(6.5, 3.5),
    cmap='dark24',
    title='COVID-19 Total Cases in Brazil by State',
    bar_label_size=7,
    tick_label_size=7,
    period_label_size=16,
    fig=None)
HTML(bcr_html)
```
